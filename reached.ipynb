```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import joblib  # Importing joblib directly

data_raw = pd.read_csv('data_raw.csv')
data_raw = data_raw.drop(['Unnamed: 0', 'shares','saves'], axis=1)

df = pd.get_dummies(data_raw, columns=['image_type'], prefix='', prefix_sep='')

X = df[['Followers', 'Time', 'envelope', 'pillow']]
y = df['Likes']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

# Save the model
joblib.dump(model, 'linear_regression_model_likes.pkl')

# Load the model from the file
loaded_model = joblib.load('linear_regression_model_likes.pkl')

# Example of using the loaded model for prediction
# Assuming Followers=1000, Time=10, envelope=0, pillow=1
input_data = pd.DataFrame([[1000, 10, True, False]], columns=['Followers', 'Time', 'envelope', 'pillow'])
predicted_likes = loaded_model.predict(input_data)
print("Predicted likes:", predicted_likes[0])
```
